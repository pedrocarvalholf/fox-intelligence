# Saving images from URL columns

```python
import urllib.request

def saveImage(i, u, file_path):
    
    filename = 'image-{}.jpg'.format(i)
    full_path = '{}{}'.format(file_path, filename)
    urllib.request.urlretrieve(url, full_path)
    
    print('{} saved.'.format(filename))
    return None

#set constants

FILENAME = 'technical_test_external_source_extract.csv'
FILE_PATH = 'images external extract/'

for i, url in enumerate(dataExternal['icon']):
    saveImage(i, url[0], FILE_PATH)
    

    

```
